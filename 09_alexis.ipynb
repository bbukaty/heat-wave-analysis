```python
import matplotlib.pyplot as plt

from utils import *

file_path = 'external/robomussels/alexis_data/fixed-data.csv'
data = pd.read_csv(file_path, parse_dates=['Timestamp'], date_format='%Y-%m-%d %H:%M:%S')

# had a bit of difficulty with the lack of timezones in the data, have to manually indicate no daylight savings time here
data['Timestamp'] = pd.DatetimeIndex(data['Timestamp']).tz_localize('EST', ambiguous=False)
data.set_index('Timestamp', inplace=True)
data.sort_index(inplace=True)
```

```python
plt.figure(figsize=(40, 6))
plt.plot(data.index, data)
```
